```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
data_dir ='data/argument_extraction_corpus_2015/annotated-phrases-by-topic/'
```

```python
dp = pd.read_csv(os.path.join(data_dir, 'dp-slider-means.csv'))
evo= pd.read_csv(os.path.join(data_dir, 'evo-slider-means.csv'))
gc = pd.read_csv(os.path.join(data_dir, 'gc-slider-means.csv'))
gm = pd.read_csv(os.path.join(data_dir, 'gm-slider-means.csv'))
```

# distribution of connectives

```python
evo_no_con = evo[evo['Connective.x']=='no_connective']['GoodSliderMean'].to_list()
evo_con = evo[evo['Connective.x']!='no_connective']['GoodSliderMean'].to_list()
evo_if = evo[evo['Connective.x']=='if']['GoodSliderMean'].to_list()
evo_but = evo[evo['Connective.x']=='but']['GoodSliderMean'].to_list()
evo_so = evo[evo['Connective.x']=='so']['GoodSliderMean'].to_list()
evo_first = evo[evo['Connective.x']=='first']['GoodSliderMean'].to_list()
```

```python
len(evo_if),len(evo_but),len(evo_so),len(evo_first),len(evo_con), len(evo_no_con)
```

```python
dp_no_con = dp[dp['Connective.x']=='no_connective']['GoodSliderMean'].to_list()
dp_con = dp[dp['Connective.x']!='no_connective']['GoodSliderMean'].to_list()
dp_if = dp[dp['Connective.x']=='if']['GoodSliderMean'].to_list()
dp_but = dp[dp['Connective.x']=='but']['GoodSliderMean'].to_list()
dp_so = dp[dp['Connective.x']=='so']['GoodSliderMean'].to_list()
dp_first = dp[dp['Connective.x']=='first']['GoodSliderMean'].to_list()
```

```python
len(dp_if),len(dp_but),len(dp_so),len(dp_first),len(dp_con), len(dp_no_con)
```

```python
gc_no_con = gc[gc['Connective.x']=='no_connective']['GoodSliderMean'].to_list()
gc_con = gc[gc['Connective.x']!='no_connective']['GoodSliderMean'].to_list()
gc_if = gc[gc['Connective.x']=='if']['GoodSliderMean'].to_list()
gc_but = gc[gc['Connective.x']=='but']['GoodSliderMean'].to_list()
gc_so = gc[gc['Connective.x']=='so']['GoodSliderMean'].to_list()
gc_first = gc[gc['Connective.x']=='first']['GoodSliderMean'].to_list()
len(gc_if)+len(gc_but)+len(gc_so)+len(gc_first)==len(gc_con)
```

```python
len(gc_if), len(gc_but),len(gc_so),len(gc_first), len(gc_con), len(gc_no_con)
```

```python
gm_no_con = gm[gm['Connective.x']=='no_connective']['GoodSliderMean'].to_list()
gm_con = gm[gm['Connective.x']!='no_connective']['GoodSliderMean'].to_list()
gm_if = gm[gm['Connective.x']=='if']['GoodSliderMean'].to_list()
gm_but = gm[gm['Connective.x']=='but']['GoodSliderMean'].to_list()
gm_so = gm[gm['Connective.x']=='so']['GoodSliderMean'].to_list()
gm_first = gm[gm['Connective.x']=='first']['GoodSliderMean'].to_list()
len(gm_if)+len(gm_but)+len(gm_so)+len(gm_first)==len(gm_con)
```

```python
gm_agree = gm[gm['Connective.x']=='i agree that']['GoodSliderMean'].to_list()
gm_no_con += gm_agree
```

```python
len(gm_if),len(gm_but),len(gm_so),len(gm_first),len(gm_con), len(gm_no_con)
```

```python
avg_no_cons = np.mean(dp_no_con+evo_no_con+ gc_no_con+gm_no_con)
avg_cons = np.mean(dp_con + evo_con+gc_con+gm_con)
avg_ifs = np.mean(dp_if + evo_if+ gc_if + gm_if)
avg_buts = np.mean(dp_but + evo_but+ gc_but+ gm_but)
avg_sos = np.mean( dp_so + evo_so + gc_so + gm_so)
avg_firsts = np.mean(dp_first + evo_first + gc_first + gm_first)
```

```python
non_cons = [np.mean(dp_no_con),np.mean(evo_no_con), np.mean(gc_no_con), np.mean(gm_no_con), avg_no_cons]
cons = [np.mean(dp_con), np.mean(evo_con), np.mean(gc_con), np.mean(gm_con), avg_cons]
ifs =  [np.mean(dp_if), np.mean(evo_if), np.mean(gc_if), np.mean(gm_if), avg_ifs]
buts = [np.mean(dp_but), np.mean(evo_but), np.mean(gc_but), np.mean(gm_but), avg_buts]
sos = [np.mean(dp_so), np.mean(evo_so), np.mean(gc_so), np.mean(gm_so), avg_sos]
firsts = [np.mean(dp_first), np.mean(evo_first), np.mean(gc_first), np.mean(gm_first), avg_firsts]
```

```python
avg_cons-avg_no_cons
```

```python
avg_ifs - avg_no_cons, avg_buts-avg_no_cons, avg_sos - avg_no_cons, avg_firsts-avg_no_cons
```

```python
xs= ['Death Peanlty', 'Evolution', 'Gun Control', 'Gay Marriage', 'AVG_OVERALL']
plt.figure(figsize=(25, 15))
colors = ['#E69F00', '#56B4E9', '#F0E442', '#009E73', '#D55E00', 'margenta']

# fig, ax= plt.subplots()
plt.plot(xs, non_cons, marker='o', linewidth=4, markersize=12, label='No Connectives', color=colors[0])
plt.plot(xs, cons, marker='o',label='All Connectives', color=colors[1],linewidth=4, markersize=12)
plt.plot(xs, ifs,marker='o', label='IF',linestyle='dashed', linewidth=4, markersize=12, color=colors[2])
plt.plot(xs, buts, marker='o',label='BUT', linestyle='dashed', linewidth=4, markersize=12,color=colors[3])
plt.plot(xs, sos,marker='o', label='SO', linestyle='dashed', linewidth=4, markersize=12,color=colors[4])
plt.plot(xs, firsts, marker='o',label='FIRST',linestyle='dashed', linewidth=4, markersize=12, color='magenta')
plt.legend(fontsize=26)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.xlabel('Topic', fontsize=30)
plt.ylabel('Mean Quality Score',fontsize=30)
plt.title('Mean Argument Quality Scores with/without Connectives', fontsize=36)
plt.savefig('connective.png')
```

```python
sents_dp = dp['Phrase.x'].to_list()
sents_evo = evo['Phrase.x'].to_list()
sents_gc= gc['Phrase.x'].to_list()
sents_gm = gm['Phrase.x'].to_list()
```

```python
all_sents = sents_dp + sents_evo + sents_gc +sents_gm
```

```python
tokenized_sents = [sent.split(' ') for sent in all_sents]
```

```python
with open('sentences_argument_mining.txt', 'w') as file:
    for sent in all_sents:
        file.write(sent+'\n')
```

```python
from collections import defaultdict
stats = defaultdict(int)
freqs = []
for tokens in tokenized_sents:
    stats[len(tokens)]+=1
    freqs.append(len(tokens))
    if len(tokens)<5:
        print(' '.join(tokens))
    
```
